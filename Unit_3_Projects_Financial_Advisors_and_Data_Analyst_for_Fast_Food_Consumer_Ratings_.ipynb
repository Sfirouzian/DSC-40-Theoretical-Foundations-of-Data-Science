**Part 1: Fast Food Restaurant Customer Satisfaction Index**

The American Customer Satisfaction Index (ACSI) scores on a 0-100 scale at the national level, with 100 being the highest and best possible score. The ACSI is an economic indicator based on modeling customer evaluations of the quality of goods and services purchased in the United States and produced by domestic and foreign firms with substantial U.S. market shares. The scores from 2007 to 2020 of the top 10 fast food companies are listed in an excel file on GitHub: 'https://raw.githubusercontent.com/Sfirouzian/Datasets/main/FastFood.csv'

```python
##Loading the Fast Food Data into Google Colab
url = 'https://raw.githubusercontent.com/Sfirouzian/Datasets/main/FastFood.csv'
FastFood = pd.read_csv(url)
FastFood.head()
```

```python
#Numeical Analysis of the fast food company data
df = pd.DataFrame(FastFood)

# Calculate descriptive statistics for mean, std, median, and IQR
statistics = df.describe().loc[['mean', 'std', '50%']]
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Add IQR to the statistics DataFrame
statistics.loc['IQR'] = IQR

# Rename '50%' to 'Median'
statistics = statistics.rename(index={'50%': 'Median'})

# Display the result
print(statistics)
```
